```python
import numpy as np

sims = 1000000

TaskA = np.random.uniform(1,20,sims)
TaskB = np.random.uniform(5,25,sims)

duration = TaskB - TaskA

print(duration)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.hist(duration, density=True, bins=50, edgecolor='black')
plt.axvline(9, color='red', linestyle='dashed', linewidth=1)
plt.title('Distribution of Task Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
print((duration > 9).sum()/sims)
```
